```
from pandas import read_csv
from statsmodels.tsa.stattools import adfuller
df = read_csv('stationary.csv', header=0, index_col=0, squeeze=True)
```

```
df = df.drop(columns=['Unnamed: 49','Unnamed: 50',	'Unnamed: 51',	'Unnamed: 52',	'Unnamed: 53',	'Unnamed: 54',	'Unnamed: 55',	'Unnamed: 56',	'Unnamed: 57',	'Unnamed: 58',	'Unnamed: 59',	'Unnamed: 60'])
```

```
IDs = ['POSNU009B003','POSNK010B053','POSNEQK200P001','POSNU022B014','POSNU039B005','POSNU039B007','POSNU043B020','POSNU161B004',
'POSNU061B005','POSNU153B002','POSNU067B002','POSNU067B004','POSNU106B004','POSNU106B040','POSNU032B003','POSNU032B005','POSNU030B003','POSNU030B022','POSNU080B007','POSNU189B015','POSNU069B002','POSNU112B003','POSNU079B004','POSNU224B002']
```

```
count = 0
while count<24:
  print('*****' + IDs[count]+ '*****')
  X = df[f'Return{count+1}'].values
  X = [a for a in X if str(a)!='nan']
  Y = df[f'MarketPrice{count+1}'].values
  Y = [a for a in Y if str(a)!='nan']
  result1 = adfuller(X, regression = 'nc')
  result2 = adfuller(X, regression = 'c')
  result3 = adfuller(X, regression = 'ct')
  result4 = adfuller(Y, regression = 'nc')
  result5 = adfuller(Y, regression = 'c')
  result6 = adfuller(Y, regression = 'ct')

  print('Return in Base Currency with No Drift')
  print('ADF Statistic: %f' % result1[0])
  print('Critical Values:')
  for key, value in result1[4].items():
    print('\t%s: %.3f' % (key, value))
  Returnlevels1 = []
  for key, value in result1[4].items():
    if result1[0]< value:
      Returnlevels1.append(key)
  if Returnlevels1 != []:
    print('The time series of return in base currency is stationary at significance level(s):')
    print(Returnlevels1)
  else:
    print('The time series of return in base currency is not stationary.')
  print('\n')
  
  print('Return in Base Currency with Drift')
  print('ADF Statistic: %f' % result2[0])
  print('Critical Values:')
  for key, value in result2[4].items():
    print('\t%s: %.3f' % (key, value))
  Returnlevels2 = []
  for key, value in result2[4].items():
    if result2[0]< value:
      Returnlevels2.append(key)
  if Returnlevels2 != []:
    print('The time series of return in base currency is stationary at significance level(s):')
    print(Returnlevels2)
  else:
    print('The time series of return in base currency is not stationary.')
  print('\n')

  print('Return in Base Currency with Drift and Trend')
  print('ADF Statistic: %f' % result3[0])
  print('Critical Values:')
  for key, value in result3[4].items():
    print('\t%s: %.3f' % (key, value))
  Returnlevels3 = []
  for key, value in result3[4].items():
    if result3[0]< value:
      Returnlevels3.append(key)
  if Returnlevels3 != []:
    print('The time series of return in base currency is stationary at significance level(s):')
    print(Returnlevels3)
  else:
    print('The time series of return in base currency is not stationary.')
  print('\n')

  print('Return in Market Price with no Drift')
  print('ADF Statistic: %f' % result4[0])
  print('Critical Values:')
  for key, value in result4[4].items():
    print('\t%s: %.3f' % (key, value))
  Returnlevels4 = []
  for key, value in result4[4].items():
    if result4[0]< value:
      Returnlevels4.append(key)
  if Returnlevels4 != []:
    print('The time series of market price in base currency is stationary at significance level(s):')
    print(Returnlevels4)
  else:
    print('The time series of market price in base currency is not stationary.')
  print('\n')

  print('Return in Market Price with Drift')
  print('ADF Statistic: %f' % result5[0])
  print('Critical Values:')
  for key, value in result5[4].items():
    print('\t%s: %.3f' % (key, value))
  Returnlevels5 = []
  for key, value in result5[4].items():
    if result5[0]< value:
      Returnlevels5.append(key)
  if Returnlevels5 != []:
    print('The time series of market price in base currency is stationary at significance level(s):')
    print(Returnlevels5)
  else:
    print('The time series of market price in base currency is not stationary.')
  print('\n')

  print('Return in Market Price with Drift and Trend')
  print('ADF Statistic: %f' % result6[0])
  print('Critical Values:')
  for key, value in result6[4].items():
    print('\t%s: %.3f' % (key, value))
  Returnlevels6 = []
  for key, value in result6[4].items():
    if result6[0]< value:
      Returnlevels6.append(key)
  if Returnlevels6 != []:
    print('The time series of market price in base currency is stationary at significance level(s):')
    print(Returnlevels6)
  else:
    print('The time series of market price in base currency is not stationary.')

  print('----------------------------------------')
  print('\n')
  count+=1
```
